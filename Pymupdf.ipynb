```python
import fitz
```

```python
doc = fitz.open('Canada.pdf')
```

```python
#Print number of Pages
print(doc.page_count)
```

```python
#print metadata

print(doc.metadata)
```

```python
#print text from pages

# for page in doc.pages():
#     print(page.get_text())
```

```python
#print text from first page

page = doc.load_page(0)
text = page.get_text()
print(text.replace("\n", " ")) #("/t", " ")
```

```python
#Send text to a .txt file

file = open("Canada.txt", "w")
file.write(text.replace("\t", " "))
file.close()
```

```python
#Extract a page from pdf and create another pdf file (Split PDF)

PDF = fitz.open("Canada.pdf")
PDF2 = fitz.open()
PDF2.insert_pdf(PDF, from_page=2, to_page = 2)
PDF2.save ("output.pdf")
```

```python
#Merge the pdf
PDF = fitz.open("Canada.pdf")
num_pages = PDF.page_count
PDF2 = fitz.open("output.pdf")
PDF2.insert_pdf(PDF, from_page=0, to_page=num_pages-1)
PDF2.save ("merged.pdf")

```

```python
#Save page image

page = doc.load_page(2)
pix = page.get_pixmap()
pix.save(f"page_{page.number}.png")
```

```python
#Save all links

for page in doc.pages():
    links = page.get_links()
    for link in links:
        print(link)
```
